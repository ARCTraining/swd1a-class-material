```python
# import your libraries
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
# import your data

folder_path = "/content/swc-python/data/"  # remember to add "/" at the end

file_name = "inflammation-01.csv"

print(folder_path + file_name)

data = np.loadtxt(fname=folder_path + file_name, delimiter=",")
```

```python
# make a wildcard filename

wildcard_file = "inflammation*.csv"
print(folder_path + wildcard_file)
```

```python
glob.glob(folder_path + wildcard_file)
```

```python
sorted_filenames = sorted(glob.glob(folder_path + wildcard_file))
```

```python
sorted_filenames[0]
```

```python
# lets make a plot!

filename = sorted_filenames[0]

data = np.loadtxt(fname=filename, delimiter=",")

# set up the figure
fig = plt.figure(figsize=(5.0, 7.0), )

# set up axes
axes_1 = fig.add_subplot(2, 1, 1)
axes_2 = fig.add_subplot(2, 1, 2)

# plot some data

axes_1.plot(np.mean(data, axis=0))

heatmap = axes_2.imshow(data, aspect="auto")

# add some labels
axes_1.set_xlabel("Days")
axes_2.set_xlabel("Days")

axes_1.set_ylabel("Mean inflammation")
axes_2.set_ylabel("Patient ID")
```

```python
figures_path = "/content/swc-python/plots/" # remember to create a folder called "plots" inside content/swc-python/

i = 1
for filename in sorted_filenames:
    print(filename)
    data = np.loadtxt(fname=filename, delimiter=",")

    # set up the figure
    fig = plt.figure(figsize=(5.0, 7.0), )

    # set up axes
    axes_1 = fig.add_subplot(2, 1, 1)
    axes_2 = fig.add_subplot(2, 1, 2)

    # plot some data

    axes_1.plot(np.mean(data, axis=0))

    heatmap = axes_2.imshow(data, aspect="auto")

    # add some labels
    axes_1.set_xlabel("Days")
    axes_2.set_xlabel("Days")

    axes_1.set_ylabel("Mean inflammation")
    axes_2.set_ylabel("Patient ID")
    plt.savefig(figures_path + "inflam_plot" + str(i) +".png")
    i = i + 1

```

```python
composite_data = np.zeros((60, 40))

for filename in sorted_filenames:
    print(filename)
    data = np.loadtxt(fname=filename, delimiter=",")
    composite_data = composite_data + data

average_data = composite_data/len(sorted_filenames)
```
