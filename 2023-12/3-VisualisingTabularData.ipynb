### Loading data into Python

To begin processing the clinical trial inflammation data, we need to load it into Python. We can do that using a library called NumPy, which stands for Numerical Python.

```python
# Load the numpy library
import numpy
```

```python
# Ask the library to read our data file for us
data = numpy.loadtxt(fname='/content/swc-python/data/inflammation-01.csv', delimiter=',')
```

### Heat Map

```python
# First, we will import the pyplot module from matplotlib
# Also: use an alias (plt) for simplicity
import matplotlib.pyplot as plt
```

```python
# create a heat map of our data
image = plt.imshow(data)

# display the plot
plt.show()
```

```python
# average inflammation over time
ave_inflammation = numpy.mean(data, axis=0)
```

```python
# line graph of average inflammation over time
plt.plot(ave_inflammation)

# display the plot
plt.show()
```

```python
# line graph of max inflammation over time
max_plot = plt.plot(numpy.amax(data, axis=0))

# display the plot
plt.show()
```

```python
# min inflammation over time
min_inflammation = numpy.amin(data, axis=0)

# line graph of min inflammation over time
plt.plot(min_inflammation)

# display the plot
plt.show()
```
